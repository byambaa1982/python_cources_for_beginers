# Merge Tables


```python
import pandas as pd
```

```python
data = {
    'Name': ['Emma', 'Oliver', 'Ava', 'Lucas', 'Mia'],
    'Age': [10, 12, None, 10, 12],
    'Grade': [85, None, 88, 92, 95]
}
df = pd.DataFrame(data)
df
```

```python
df_clean = df.fillna(5.0)
df_clean
```

```python
data1 = {
    'Name': ['Emma', 'Oliver', 'Ava'],
    'Grade': [45, 90, 21],
    'id': [1, 2,3]
}

data2 = {
    'Name': ['Emma', 'Oliver', 'Lucas'],
    'Age': [10, 12, 10],
    'id': [1, 2,4]
}
```

```python
df1 = pd.DataFrame(data1)
df1
```

```python
df2 = pd.DataFrame(data2)
df2
```

```python
inner_merged_df = pd.merge(df1, df2, on = ['id','Name'], how = 'inner')
inner_merged_df
```

```python
outer_merged_df = pd.merge(df1, df2, on = ['id','Name'], how = 'outer')
outer_merged_df
```

```python
df_g= pd.merge(df1, df2, on = ['id','Name'], how = 'left')
```

```python
df_g
```

```python
df_tt = pd.merge(df1, df2, on = ['id','Name'], how = 'right')
```
