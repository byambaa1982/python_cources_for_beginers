```python
import pandas as pd

# Create a DataFrame with 10 rows
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Ella', 'Frank', 'Grace', 'Hannah', 'Ian', 'Jack'],
    'Favorite Animal': ['Panda', 'Elephant', 'Giraffe', 'Panda', 'Koala', 'Lion', 'Dolphin', 'Panda', 'Tiger', 'Penguin'],
    'Age': [10, 12, 11, 10, 9, 13, 11, 10, 12, 9],
    'Score': [85, 92, 88, 76, 95, 89, 91, 87, 84, 90],
    'Favorite Color': ['Red', 'Blue', 'Green', 'Yellow', 'Purple', 'Orange', 'Pink', 'Teal', 'Brown', 'Black']
}

df = pd.DataFrame(data)

df
```

```python
df[df['Age']==10]
```

```python
df[(df['Age']>=10) & (df['Age']<=11)]
```

```python
df[["Name", "Age", "Score"]]
```

```python
name_age_score = df[["Name", "Age", "Score"]]
```

```python
name_age_score
```

```python
average_score = df["Score"].mean()
print(f"The average age is {average_score}")
max_score = df["Score"].max()
print(f"The average age is {max_score}")
minimum_score = df["Score"].min()
print(f"The average age is {minimum_score}")
```

```python
df_sorted = df.sort_values(by= 'Score')
df_sorted
```

```python
def find_average(col):
  average_score = df[col].mean()
  print(f"The average {col} is {average_score}")
```

```python
find_average('Age')
```

```python
find_average('Score')
```

```python
def find_maximum(col):
  average_score = df[col].max()
  print(f"The maximum {col} is {max_score}")
```

```python
find_maximum('Age')
find_maximum('Score')
```
