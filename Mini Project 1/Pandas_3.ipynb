```python
joined_df['Average_Grade'] = (joined_df['Math_Grade']+joined_df['Science_Grade'])/2
joined_df
```

```python
def make_grade(x):
  if x >= 90 and x <=100:
    print('A')
    return "A"
  elif x >= 80 and x < 90:
    return "B"
    print('B')
  elif x >= 70 and x < 80:
    return "C"
    print('C')
  elif x >= 60 and x < 70:
    return "D"
    print('D')
  else:
    print('F')
```

```python
joined_df['Average_Grade'] = (joined_df['Math_Grade']+joined_df['Science_Grade'])/2
joined_df['Grade'] = joined_df['Average_Grade'].map(lambda x: make_grade(x))
joined_df['my_hobby'] = joined_df['hobby'].map(lambda x: "I like" + x)
joined_df
```
