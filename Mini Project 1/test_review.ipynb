```python
data1 = {
    'Name': ['Emma', 'Oliver', 'Ava'],
    'Grade': [45, 90, 21],
    'id': [1, 2,3]
}

data2 = {
    'Name': ['Emma', 'Oliver', 'Lucas'],
    'Age': [10, 12, 10],
    'id': [1, 2,4]
}
```

```python
df1 = pd.DataFrame(data1)
df1
```

```python
df2 = pd.DataFrame(data2)
df2
```

```python
inner_merged_df = pd.merge(df1, df2, on = ['id','Name'], how = 'inner')
inner_merged_df
```

```python
outer_merged_df = pd.merge(df1, df2, on = ['id','Name'], how = 'outer')
outer_merged_df
```

```python
df_l = pd.merge(df1, df2, on = ['id','Name'], how = 'left')
df_l
```
