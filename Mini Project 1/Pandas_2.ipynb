# Handling Missing Data


```python
import pandas as pd
```

```python
data = {
    'Name': ['Emma', 'Oliver', 'Ava', 'Lucas', 'Mia'],
    'Age': [10, 12, None, 10, 12],
    'Grade': [85, None, 88, 92, 95]
}
df = pd.DataFrame(data)
df
```

```python
df_clean = df.fillna(5)
df_clean
```

```python
df_clean = df.fillna(3)
df_clean
```

```python
grouped_df = df.groupby('Age').mean(numeric_only = True)
grouped_df
```

```python
data1 = {
    'Name': ['Emma', 'Oliver', 'Ava'],
    'Grade': [45, 90, 88],
    'id': [1,2,3]
}

data2 = {
    'Name': ['Emma', 'Oliver', 'Lucas'],
    'Age': [10, 12, 10],
    'id': [1, 2,4]
}
```

```python
df1 = pd.DataFrame(data1)
df1
```

```python
df2 = pd.DataFrame(data2)
df2
```

```python
outer_merged_df = pd.merge(df1, df2, on = 'Name', how = 'outer')
outer_merged_df
```

```python
left_df = pd.merge(df1, df2, on = 'Name', how = 'left')
left_df
```

```python
right_df = pd.merge(df1, df2, on = 'Name', how = 'right')
right_df
```
