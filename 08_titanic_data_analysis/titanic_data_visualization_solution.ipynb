```
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('titanic.csv')
df.head()
```

# Data Explore

```
# Check for missing values
df.isnull().sum()
```

```
df.isnull().sum().sum()
```

```
df.describe()
```

# Charts

# 1. Bar Chart - Gender Distribution

```
plt.figure(figsize=(8, 5))
sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')
plt.show()
```

```
# 1. Bar Chart - Passenger Class Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='Pclass', data=df, hue='Pclass', palette='Set2', legend=False)
plt.title('Passenger Class Distribution')
plt.show()
```

## Task: Create a bar chart that shows the distribution of survivors and non-survivors for each gender.

```
# Solution for Bar Chart: Gender Distribution by Survival
plt.figure(figsize=(8, 5))
sns.countplot(x='Sex', hue='Survived', data=df, palette='Set1')
plt.title('Gender Distribution by Survival')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['Did Not Survive', 'Survived'])
plt.show()
```

# 2. Pie Chart - Survival Rate

```
# 2. Pie Chart - Survival Rate
survived_counts = df['Survived'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(survived_counts, labels=['Did Not Survive', 'Survived'], autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'skyblue'])
plt.title('Survival Rate')
plt.show()
```

```
# 2. Pie Chart - Gender Proportion
gender_counts = df['Sex'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=['Male', 'Female'], autopct='%1.1f%%', startangle=90, colors=['lightblue', 'pink'])
plt.title('Gender Proportion')
plt.show()
```

## Task: Create a pie chart showing the proportion of passengers based on the port they embarked from (Embarked column).

```
# Solution for Pie Chart: Proportion of Passengers by Embarked Location
embarked_counts = df['Embarked'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(embarked_counts, labels=embarked_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Proportion of Passengers by Embarked Location')
plt.show()
```

# 3. Histogram - Age Distribution

```

plt.figure(figsize=(8, 5))
sns.histplot(df['Age'].dropna(), bins=20, kde=False)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()
```

```
# 3. Histogram - Fare Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Fare'], bins=30, kde=False)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.show()
```

## Task: Create a histogram showing the age distribution of passengers.

```
# Solution for Histogram: Age Distribution of Passengers
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'].dropna(), bins=20, kde=False)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

# 4. Box Plot - Fare by Passenger Class

```

plt.figure(figsize=(8, 5))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare by Passenger Class')
plt.show()
```

```
# 4. Box Plot - Age by Survival Status
plt.figure(figsize=(8, 5))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age Distribution by Survival Status')
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.show()
```

# 5. Scatter Plot - Age vs Fare

```

plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Age vs Fare')
plt.show()
```

```
# 5. Scatter Plot - Age vs SibSp (Number of Siblings/Spouses Aboard)
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='SibSp', data=df)
plt.title('Age vs Number of Siblings/Spouses Aboard')
plt.show()
```

# 6. Line Chart - Cumulative Survival by Age

```

df_sorted = df[['Age', 'Survived']].sort_values(by='Age').dropna()
df_sorted['Cumulative_Survived'] = df_sorted['Survived'].cumsum()
plt.figure(figsize=(8, 5))
plt.plot(df_sorted['Age'], df_sorted['Cumulative_Survived'])
plt.title('Cumulative Survival by Age')
plt.xlabel('Age')
plt.ylabel('Cumulative Survival')
plt.show()
```

```
# 6. Line Chart - Cumulative Count of Passengers by Fare
df_sorted_fare = df[['Fare']].sort_values(by='Fare').dropna()
df_sorted_fare['Cumulative_Count'] = df_sorted_fare.index + 1
plt.figure(figsize=(8, 5))
plt.plot(df_sorted_fare['Fare'], df_sorted_fare['Cumulative_Count'])
plt.title('Cumulative Count of Passengers by Fare')
plt.xlabel('Fare')
plt.ylabel('Cumulative Count')
plt.show()
```

# 7. Stacked Bar Chart - Survival by Class

```

survived_class = df.groupby(['Pclass', 'Survived']).size().unstack()
survived_class.plot(kind='bar', stacked=True, figsize=(8, 5), color=['lightcoral', 'skyblue'])
plt.title('Survival by Class')
plt.ylabel('Count')
plt.show()
```

```
# 7. Stacked Bar Chart - Gender by Survival
gender_survived = df.groupby(['Sex', 'Survived']).size().unstack()
gender_survived.plot(kind='bar', stacked=True, figsize=(8, 5), color=['lightcoral', 'skyblue'])
plt.title('Survival by Gender')
plt.ylabel('Count')
plt.show()
```

# 8. Heatmap - Correlation Matrix

```

plt.figure(figsize=(8, 5))
corr_matrix = df[['Age', 'Fare', 'Pclass', 'SibSp', 'Parch', 'Survived']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# 9. Violin Plot - Fare Distribution by Class (Advanced)

```

plt.figure(figsize=(8, 5))
sns.violinplot(x='Pclass', y='Fare', data=df)
plt.title('Fare Distribution by Class')
plt.show()
```

```
# 9. Violin Plot - Age Distribution by Survival Status (Advanced)
plt.figure(figsize=(8, 5))
sns.violinplot(x='Survived', y='Age', data=df)
plt.title('Age Distribution by Survival Status')
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.show()
```
