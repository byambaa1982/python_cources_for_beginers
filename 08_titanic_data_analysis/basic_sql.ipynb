```
# Finding the Average Age
query = '''
select AVG(Age) as Average_age from titanic
'''

df = pd.read_sql(query, conn)

df
```

# Task 11: Find the Most Expensive Fare Paid by Survivors

Write an SQL query to find the highest fare paid by passengers who survived, along with the name(s) of the passengers.

```
query = '''
select Name, Age,
 CASE
  when Age between 0 and 12 then 'Child'
  when Age between 13 and 19 then 'Teen'
  when Age between 20 and 59 then 'Adult'
  ELSE 'Senior'
 END AS Age_Group
from titanic
'''

df = pd.read_sql(query, conn)

df
```

```
query = '''
select Name, Survived,
 CASE
  when Survived ==1 then 'Alive'
  ELSE 'Dead'
 END AS Alive_or_Dead
from titanic
'''

df = pd.read_sql(query, conn)

df
```
