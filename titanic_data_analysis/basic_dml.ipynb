```
import pandas as pd
import sqlite3
```


# SQL

- Step 1. Download titanic.sqlite

- step 2. Upload file into notebook

- step 3. Run the first two cells

```
# Connect to the SQLite database
conn = sqlite3.connect('titanic.sqlite')

# Read the 'titanic' table into a pandas DataFrame
df = pd.read_sql('SELECT * FROM titanic', conn)
df
```

# Use 'PassengerId', 'Name', 'Sex', 'Age' columns and save dataframe into database as passenger_info

```
# 1. Passenger Info Table
passenger_info = df[['PassengerId', 'Name', 'Sex', 'Age']]
passenger_info.to_sql('passenger_info', conn, if_exists='replace', index=False)

# 2. Ticket Info Table
ticket_info = df[['PassengerId', 'Ticket', 'Fare', 'Pclass']]
ticket_info.to_sql('ticket_info', conn, if_exists='replace', index=False)

# 3. Survival Status Table
survival_status = df[['PassengerId', 'Survived', 'Parch', 'SibSp']]
survival_status.to_sql('survival_status', conn, if_exists='replace', index=False)
```

# Task 1: Update Task
Objective:

- Update the age of a passenger based on their PassengerId.

Instructions:

- Update the age of the passenger with PassengerId = 5 in the passenger_info table to 40.

```

update_query = '''
update
  passenger_info
set
  Age = 40
where
  PassengerId = 5
'''
cursor = conn.cursor()
cursor.execute(update_query)
conn.commit()

q = '''
select
PassengerId, Name, Sex, Age
from
    passenger_info
limit
  10
'''

p_info = pd.read_sql(q, conn)
p_info
```

# Task 2: Delete Task
Objective:

- Delete a passenger's ticket information based on their PassengerId.

Instructions:

- Delete the record of the passenger with PassengerId = 8 from the ticket_info table.

```

delete_query = '''
delete from
  passenger_info
where
  PassengerId = 8
'''
cursor = conn.cursor()
cursor.execute(delete_query)
conn.commit()

q = '''
select
PassengerId, Name, Sex, Age
from
    passenger_info
limit
  10
'''

p_info = pd.read_sql(q, conn)
p_info
```

# Task 3: Insert Task
Objective:

- Insert a new passenger's survival status into the survival_status table.

Instructions:

- Insert a new record with the following details into the survival_status table:
- PassengerId = 891
- Survived = 1
- Parch = 0
- SibSp = 1

```
insert_query = '''
insert into
  survival_status (PassengerId, Survived, Parch, SibSp)
values
  (891, 1, 0, 1)
'''
cursor = conn.cursor()
cursor.execute(insert_query)
conn.commit()


select_query = '''
select
  PassengerId, Survived, Parch, SibSp
from
  survival_status
where
  PassengerId = 891
limit
  2
'''

p_info = pd.read_sql(select_query, conn)
p_info
```

# Task 4: Update Multiple Columns
Objective:

- Update multiple columns for a specific passenger in the ticket_info table.

Instructions:

- Update the Fare to 150.0 and Pclass to 1 for the passenger with PassengerId = 4 in the ticket_info table.

```
update_query = '''
update
  ticket_info
set
   Fare = 150,
   Pclass = 1
where
  PassengerId = 4
'''
cursor = conn.cursor()
cursor.execute(update_query)
conn.commit()


select_query = '''
select
  PassengerId, Ticket, Fare, Pclass
from
  ticket_info
where
  PassengerId = 4
'''

p_info = pd.read_sql(select_query, conn)
p_info
```

# Task 5: Delete Based on Condition
Objective:

- Delete records where a condition is met in the survival_status table.

Instructions:

- Delete all passengers from the survival_status table where Survived = 0 (i.e., those who did not survive).

```
delete_query = '''
delete from
  survival_status
where
  Survived = 0
'''
cursor = conn.cursor()
cursor.execute(delete_query)
conn.commit()


select_query = '''
select
  PassengerId, Survived
from
  survival_status
where
  PassengerId = 891
'''

p_info = pd.read_sql(select_query, conn)
p_info
```
