```
query = '''
select Name, Sex, Age from titanic where Sex = "male" order by Age desc limit 10
'''

df = pd.read_sql(query, conn)

df
```

```
query = '''
select Survived, count(*) as Count from titanic GROUP BY Survived
'''

df = pd.read_sql(query, conn)

df
```

```
# Finding the Average Age
query = '''
select AVG(Age) as Average_age from titanic
'''

df = pd.read_sql(query, conn)

df
```

```
query = '''
select Embarked, count(*) as Count from titanic GROUP BY Embarked
'''

df = pd.read_sql(query, conn)

df
```

```
query = '''
select Pclass, count(*) passengers from titanic group by pclass
'''
df = pd.read_sql(query, conn)

df
```

```
query = '''
select MAX(Fare) Most_Fare from titanic
'''
df = pd.read_sql(query, conn)

df
```

```
query = '''
select MIN(Fare) Least_Fare from titanic
'''
df = pd.read_sql(query, conn)

df
```

```
query = '''
select max(Age) Age, Name from titanic
'''
df = pd.read_sql(query, conn)

df
```

```
query = '''
select min(Age) Age, Name from titanic
'''
df = pd.read_sql(query, conn)

df
```
