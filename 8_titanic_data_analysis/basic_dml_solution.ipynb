```
import pandas as pd
import sqlite3
```


# SQL

- Step 1. Download titanic.sqlite

- step 2. Upload file into notebook

- step 3. Run the first two cells

```
# Connect to the SQLite database
conn = sqlite3.connect('titanic.sqlite')

# Read the 'titanic' table into a pandas DataFrame
df = pd.read_sql('SELECT * FROM titanic', conn)
df
```

# Use 'PassengerId', 'Name', 'Sex', 'Age' columns and save dataframe into database as passenger_info

```
# 1. Passenger Info Table
passenger_info = df[['PassengerId', 'Name', 'Sex', 'Age']]
passenger_info.to_sql('passenger_info', conn, if_exists='replace', index=False)

# 2. Ticket Info Table
ticket_info = df[['PassengerId', 'Ticket', 'Fare', 'Pclass']]
ticket_info.to_sql('ticket_info', conn, if_exists='replace', index=False)

# 3. Survival Status Table
survival_status = df[['PassengerId', 'Survived', 'Parch', 'SibSp']]
survival_status.to_sql('survival_status', conn, if_exists='replace', index=False)
```

# Task 1: Update Task
Objective:

- Update the age of a passenger based on their PassengerId.

Instructions:

- Update the age of the passenger with PassengerId = 5 in the passenger_info table to 40.

```
update_query = '''
UPDATE passenger_info
SET Age = 40
WHERE PassengerId = 5
'''
cursor = conn.cursor()
cursor.execute(update_query)
conn.commit()

# Verify the update
q = '''
SELECT PassengerId, Name, Age
FROM passenger_info
WHERE PassengerId = 5
'''
updated_df = pd.read_sql(q, conn)
print(updated_df)
```

# Task 2: Delete Task
Objective:

- Delete a passenger's ticket information based on their PassengerId.

Instructions:

- Delete the record of the passenger with PassengerId = 8 from the ticket_info table.

```
delete_query = '''
DELETE FROM ticket_info
WHERE PassengerId = 8
'''
cursor = conn.cursor()
cursor.execute(delete_query)
conn.commit()

# Verify the deletion
q = '''
SELECT PassengerId, Ticket, Fare
FROM ticket_info
WHERE PassengerId = 8
'''
remaining_df = pd.read_sql(q, conn)
remaining_df
```

# Task 3: Insert Task
Objective:

- Insert a new passenger's survival status into the survival_status table.

Instructions:

- Insert a new record with the following details into the survival_status table:
- PassengerId = 891
- Survived = 1
- Parch = 0
- SibSp = 1

```
insert_query = '''
INSERT INTO survival_status (PassengerId, Survived, Parch, SibSp)
VALUES (891, 1, 0, 1)
'''
cursor = conn.cursor()
cursor.execute(insert_query)
conn.commit()

# Verify the insertion
q = '''
SELECT PassengerId, Survived, Parch, SibSp
FROM survival_status
WHERE PassengerId = 891
'''
df = pd.read_sql(q, conn)
df
```

# Task 4: Update Multiple Columns
Objective:

- Update multiple columns for a specific passenger in the ticket_info table.

Instructions:

- Update the Fare to 150.0 and Pclass to 1 for the passenger with PassengerId = 4 in the ticket_info table.

```
update_query = '''
UPDATE ticket_info
SET Fare = 150.0, Pclass = 1
WHERE PassengerId = 4
'''
cursor = conn.cursor()
cursor.execute(update_query)
conn.commit()

# Verify the update
q = '''
SELECT PassengerId, Ticket, Fare, Pclass
FROM ticket_info
WHERE PassengerId = 4
'''
updated_df = pd.read_sql(q, conn)
updated_df
```

# Task 5: Delete Based on Condition
Objective:

- Delete records where a condition is met in the survival_status table.

Instructions:

- Delete all passengers from the survival_status table where Survived = 0 (i.e., those who did not survive).

```
delete_query = '''
DELETE FROM survival_status
WHERE Survived = 0
'''
cursor = conn.cursor()
cursor.execute(delete_query)
conn.commit()

# Verify the deletion
q = '''
SELECT PassengerId, Survived
FROM survival_status
WHERE Survived = 0
'''
remaining_df = pd.read_sql(q, conn)
remaining_df
```

# Task 6: Insert New Passenger with Complete Information
Objective:

- Insert a new passenger into all three tables: passenger_info, ticket_info, and survival_status.

Instructions:

Insert a new passenger with the following details:
- PassengerId = 892
- Name = 'Tom Hardy', Sex = 'male', Age = 37 into passenger_info
- Ticket = 'PC 17599', Fare = 75.0, Pclass = 1 into ticket_info
- Survived = 1, Parch = 0, SibSp = 0 into survival_status

```
# Insert into passenger_info
insert_passenger_info = '''
INSERT INTO passenger_info (PassengerId, Name, Sex, Age)
VALUES (892, 'Tom Hardy', 'male', 37)
'''
cursor = conn.cursor()
cursor.execute(insert_passenger_info)
conn.commit()

# Insert into ticket_info
insert_ticket_info = '''
INSERT INTO ticket_info (PassengerId, Ticket, Fare, Pclass)
VALUES (892, 'PC 17599', 75.0, 1)
'''
cursor.execute(insert_ticket_info)
conn.commit()

# Insert into survival_status
insert_survival_status = '''
INSERT INTO survival_status (PassengerId, Survived, Parch, SibSp)
VALUES (892, 1, 0, 0)
'''
cursor.execute(insert_survival_status)
conn.commit()

# Verify the insertion
q = '''
SELECT * FROM passenger_info
WHERE PassengerId = 892
'''
passenger_df = pd.read_sql(q, conn)
print(passenger_df)

q = '''
SELECT * FROM ticket_info
WHERE PassengerId = 892
'''
ticket_df = pd.read_sql(q, conn)
print(ticket_df)

q = '''
SELECT * FROM survival_status
WHERE PassengerId = 892
'''
survival_df = pd.read_sql(q, conn)
print(survival_df)
```
